This file is for merging the yellow, green, fhv and hvfhv taxi data

```python
import numpy as np
import pandas as pd
import pyarrow as pa
import pyarrow.parquet as pq
import os
```

```python
# Get list of each filename
yellow_files = [filename for filename in os.listdir('.') if filename.startswith("yellow")]
green_files = [filename for filename in os.listdir('.') if filename.startswith("green")]
fhv_files = [filename for filename in os.listdir('.') if filename.startswith("fhv_")] #Need underscore to ensure it doesn't pick up fhvhv
fhvhv_files = [filename for filename in os.listdir('.') if filename.startswith("fhvhv")]
```

```python
def merge_files(filenames, output_name):
    # merge multiple files into one parquet file
    schema = pq.ParquetFile(yellow_files[0]).schema_arrow
    with pq.ParquetWriter(output_name, schema=schema) as writer:
        for file in filenames:
            writer.write_table(pq.read_table(file, schema=schema))
```

```python
files_to_merge = [fhvhv_files] # [yellow_files, green_files, fhv_files, fhvhv_files]
output_names = ["all_fhvhv_tripdata.parquet"] # ["all_yellow_tripdata.parquet", "all_green_tripdata.parquet", "all_fhv_tripdata.parquet", "all_fhvhv_tripdata.parquet"] # start name with all to ensure 

for filenames, output_name in zip(files_to_merge, output_names):
    merge_files(filenames, output_name)
    print(output_name, "merged")
```
