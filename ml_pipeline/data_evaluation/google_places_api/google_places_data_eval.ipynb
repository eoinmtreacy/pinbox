```python
df_osm1 = pd.read_csv("../osm_places/osm_places_amenity.csv")
df_osm2 = pd.read_csv("../osm_places/osm_places_leisure.csv")
df_osm3 = pd.read_csv("../osm_places/osm_places_shop.csv")
```

```python
df_osm = pd.concat([df_osm1, df_osm2, df_osm3])
```

```python
df_joined = pd.merge(left=df_places, right=df_osm, how='left', on=['id'])
```
