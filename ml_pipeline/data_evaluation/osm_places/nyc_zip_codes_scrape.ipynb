This notebook is for scraping NYC zip codes from https://www.nycbynatives.com/nyc_info/new_york_city_zip_codes.php

```python
from bs4 import BeautifulSoup
import pandas as pd
from urllib.request import urlopen
```

```python
with urlopen("https://nycbynatives.com/nyc_info/new_york_city_zip_codes.php") as response:
    soup = BeautifulSoup(response, 'html.parser')
    result = []
    tables = soup.findAll('table')
    for table in tables:
        rows = table.find_all(['tr'])
        for row in rows:
            cols = row.find_all(['td', 'td'])
            print(cols)
            text = [col.get_text(strip=True) for col in cols]
            print(len(text))
            result.append(text)
    
    df = pd.DataFrame(result)           
```

```python

df1 = df[[0, 1]]
df2 = df[[3, 4]]

df1.rename(columns={0:'zip', 1:'borough'}, inplace=True)
df2.rename(columns={3:'zip', 4:'borough'}, inplace=True)
df = pd.concat([df1, df2])
```

```python
df.to_csv("zipcodes.csv")
```
