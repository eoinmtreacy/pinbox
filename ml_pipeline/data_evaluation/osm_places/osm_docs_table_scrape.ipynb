This notebook is for scraping all the tables from the OSM map features wiki. Result is output into a CSV for easy viewing

```python
from bs4 import BeautifulSoup
import pandas as pd
from urllib.request import urlopen
```

```python
df = pd.DataFrame(columns=['Row1', 'Row2'])
```

```python
with urlopen("https://wiki.openstreetmap.org/wiki/Map_features") as response:
    soup = BeautifulSoup(response, 'html.parser')
    result = []
    tables = soup.findAll('table')
    for table in tables:
        rows = table.find_all(['tr'])
        for row in rows:
            cols = row.find_all(['td', 'td'])
            print(cols)
            first_two = [col.get_text(strip=True) for col in cols]
            print(len(first_two))
            if len(first_two)>= 2:
                result.append(first_two)
    
    df = pd.DataFrame(result)           
```
