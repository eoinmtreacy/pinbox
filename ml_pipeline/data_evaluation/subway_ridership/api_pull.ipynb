This notebook is for pulling hourly subway ridership data from NYC open data

```python
import pandas as pd
from sodapy import Socrata
import json
```

```python
f = open("keys.json")
DATANY_APP_TOKEN = json.load(f)['datany_apptoken']
f.close()
```

```python
client = Socrata("data.ny.gov", DATANY_APP_TOKEN)
# First 2000 results, returned as JSON from API / converted to Python list of
# dictionaries by sodapy.
results = client.get("wujg-7c2s", limit=5000, offset=0)
# Convert to pandas DataFrame
results_df = pd.DataFrame.from_records(results)
```
